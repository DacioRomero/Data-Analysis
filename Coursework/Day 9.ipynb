# Two normal distributions

```python
import numpy as np
import seaborn as sns
```

```python
a = np.random.normal(60, 10, 100000)
sns.distplot(a, hist=True, kde=True, bins=40)
b = np.random.normal(60, 20, 100000)
sns.distplot(b, hist=True, kde=True, bins=40)
```

```python
import scipy.stats as st
```

```python
st.norm.cdf(70, loc = 60, scale = 10) - st.norm.cdf(50, loc = 60, scale = 10)
```

```python
st.norm.cdf(80, loc = 60, scale = 20) - st.norm.cdf(40, loc = 60, scale = 20)
```

```python
def calculate_cdf(x, data):
    return np.sum(x < data)
```

```python
cdf_a = [calculate_cdf(a, r) for r in range(int(np.min(a)), int(np.max(a)))]
cdf_a = [calculate_cdf(b, r) for r in range(int(np.min(b)), int(np.max(b)))]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(int(np.min(a)), int(np.max(a))), cdf_a)
```

```python
c = (b - 60) / 20
sns.distplot(c, hist=True, kde=True, bins=40)
```
